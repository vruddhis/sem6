```python
import kagglehub
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

path = kagglehub.dataset_download("emadmakhlouf/linearly-inseperable-dataset")

df = pd.read_csv(path + '/moonDataset.csv')
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['X1'], df['X2'], df['X3'], c=df['label'])
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('X3')
ax.set_title('Linearly Inseparable Dataset')
plt.show()
```

```python
from sklearn import svm
```

```python
import numpy as np
from sklearn.model_selection import train_test_split

X = df[['X1', 'X2', 'X3']].values
y = df['label'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
clf = svm.SVC() #this is rbf kernel
clf.fit(X_train, y_train)
```

```python


from sklearn.metrics import accuracy_score

y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy on test set: {accuracy}")
```

```python
len(clf.support_vectors_)
```

```python
clf = svm.SVC(kernel = 'linear')
clf.fit(X_train, y_train)

from sklearn.metrics import accuracy_score

y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy on test set: {accuracy}")
len(clf.support_vectors_)
```

```python
clf = svm.SVC(kernel = 'sigmoid')
clf.fit(X_train, y_train)

from sklearn.metrics import accuracy_score

y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy on test set: {accuracy}")
len(clf.support_vectors_)
```

```python
clf = svm.SVC(kernel = 'poly')
clf.fit(X_train, y_train)

from sklearn.metrics import accuracy_score

y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy on test set: {accuracy}")
len(clf.support_vectors_)
```
